```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_').str.replace('(', '').str.replace(')', '').str.replace('_usd', '')
df.head(3)
```

```python
# create age groups column
labels = ['Young Adult', 'Adult', 'Middle Aged', 'Senior']
bins = [0, 24, 34, 54, 100]
df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)
df.head()
df['age_group'].value_counts().sort_index()
```

```python
# create numeric purchase frequency column
frequency_mapping ={
    "Fortnightly": 14,
    'Weekly': 7,
    'Annually': 365,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Monthly': 30,
    'Every 3 Months': 90
}
df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
df[['purchase_frequency_days', 'frequency_of_purchases']].head(20)
```

```python
# convert purchase frequency to integer type
df['purchase_frequency_days'] = df['purchase_frequency_days'].astype('Int64')
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
# verify if 'discount_applied' and 'promo_code_used' columns are identical
(df['discount_applied'] == df['promo_code_used']).all()
#drop promo_code_used column
df = df.drop(columns= ['promo_code_used'])
```
